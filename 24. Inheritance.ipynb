# Inheritance

```python
# What is inheritance?

# Inheritance is a concept in object-oriented programming where a 
# new class is created by inheriting the properties and methods of an existing class.

# The new class means the subclass or child class, while the existing class means the superclass or parent class.
```

# a. Single Inheritance

```python
# When the inheritance involves one parent class and one child class
```

```python
class Parent:
    def func1(self):
        print("This i fun1")

class Child(Parent):
    def func2(self):
        print("This is fun2")

ob = Child()
ob.func2() # from child itself
ob.func1() # can call function from parent class
```

```python
class Vehicle:
    def __init__(self, make, model, year, color):
        self.make = make
        self.model = model
        self.year = year
        self.color = color
        self.engine_running = False

    def start(self):
        self.engine_running = True
        print("Engine started.")

    def stop(self):
        self.engine_running = False
        print("Engine stopped.")

class Car(Vehicle):
    def __init__(self, make, model, year, color, num_doors):
        super().__init__(make, model, year, color)
        self.num_doors = num_doors

    def drive(self):
        print("Driving...")

my_car = Car("Toyota", "Camry", 2021, "blue", 4)  # object of child class

my_car.drive()                 # from child class itself
print(my_car.num_doors)        # from child class itself
my_car.start()                 # can call method from parent class
my_car.stop()                  # can call method from parent class
```

# b. Multiple Inheritance

```python
# Involves more than one parent class
```

```python
class Parent1:
    def func1(self):
        print("This is fun1")

class Parent2:
    def func2(self):
        print("This is fun2")

class Child(Parent1,Parent2):
    def func3(self):
        print("This is fun3")

ob = Child()
ob.func3() # from Child itself
ob.func2() # from Parent2
ob.func1() # fom Parent1
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def display(self):
        print("Name:", self.name)
        print("Salary:", self.salary)

class SalesPerson:
    def __init__(self, sales_target):
        self.sales_target = sales_target
    
    def display_sales_target(self):
        print("Sales Target:", self.sales_target)

class Manager(Employee, SalesPerson):
    def __init__(self, name, salary, sales_target):
        Employee.__init__(self, name, salary)
        SalesPerson.__init__(self, sales_target)
    
    def display_all(self):
        Employee.display(self)
        SalesPerson.display_sales_target(self)


#----------------- creating object for salesperson class --------------------

sales_person = SalesPerson(99999)
sales_person.display_sales_target()
print()

# ---------------------  creating object for manager class which derived from employye and salesperson class ----------

manager = Manager("John Smith", 50000, 1000000)

manager.display_sales_target()
manager.display()
manager.display_all()
```

# c. Multi level inheritance

```python
# Child class acts as parent class for another child class
```

```python
class Parent:
    def func1(self):
        print("This i fun1")

class Child(Parent):
    def func2(self):
        print("This i fun2")

class GrandChild(Child):
    def func3(self):
        print("This is fun3")

ob = GrandChild()
ob.func3() # from GrandChild itself
ob.func2() # from Child class
ob.func1() # from Parent class
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def display(self):
        print("Employee:", self.name)

class SalesPerson(Employee):
    def __init__(self, name, salary, sales_target):
        super().__init__(name, salary)
        self.sales_target = sales_target
    
    def display(self):
        super().display()
        print("Sales Person, Sales Target:", self.sales_target)

class SalesManager(SalesPerson):
    def __init__(self, name, salary, sales_target, team_size):
        super().__init__(name, salary, sales_target)
        self.team_size = team_size
    
    def display(self):
        super().display()
        print("Sales Manager, Team Size:", self.team_size)

manager = SalesManager("John", 50000, 1000000, 10)
manager.display() # Output: Employee: John, Sales Person, Sales Target: 1000000, Sales Manager, Team Size: 10
```

# d. Hierarchical Inheritance

```python
# When more than one derived class are created from a single base this type of inheritance is called hierarchical inheritance
```

```python
class Parent:
    def func1(self):
        print("This i fun1")

class Child1(Parent):
    def func2(self):
        print("This i fun2")

class Child2(Parent):
    def func3(self):
        print("This is fun3")

ob1 = Child1()
ob1.func2() # from Child1 class
ob1.func1() # from Parent class

ob2 = Child2()
ob2.func3() # from Child2 class
ob2.func1() # from Parent class
```

```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def display(self):
        print(f"{self.year} {self.make} {self.model}")

class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def display(self):
        super().display()
        print(f"{self.num_doors} doors")

class Truck(Vehicle):
    def __init__(self, make, model, year, payload_capacity):
        super().__init__(make, model, year)
        self.payload_capacity = payload_capacity

    def display(self):
        super().display()
        print(f"Payload capacity: {self.payload_capacity} lbs")

car = Car("Toyota", "Camry", 2021, 4)
truck = Truck("Ford", "F-150", 2022, 2000)

car.display()   # Output: 2021 Toyota Camry, 4 doors
truck.display() # Output: 2022 Ford F-150, Payload capacity: 2000 lbs
```

```python
# Parent class
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def display_details(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Gender:", self.gender)


# Child classes inheriting from Person
class Employee(Person):
    def __init__(self, name, age, gender, emp_id, salary):
        # Calling the constructor of the parent class
        Person.__init__(self, name, age, gender)
        self.emp_id = emp_id
        self.salary = salary

    def display_employee_details(self):
        print("Employee ID:", self.emp_id)
        print("Salary:", self.salary)


class Customer(Person):
    def __init__(self, name, age, gender, customer_id, address):
        # Calling the constructor of the parent class
        Person.__init__(self, name, age, gender)
        self.customer_id = customer_id
        self.address = address

    def display_customer_details(self):
        print("Customer ID:", self.customer_id)
        print("Address:", self.address)
```

# e. Hybrid Inheritance

```python
# Inheritance consisting of multiple types of inheritance is called hybrid inheritance.
```

```python
class Parent:
    def func1(self):
        print("This i fun1")

class Child1(Parent):
    def func2(self):
        print("This i fun2")

class Child2(Parent):
    def func3(self):
        print("This is fun3")

class GrandChild(Child1,Child2):
    def func4(self):
        print("This is fun4")

ob = GrandChild()
ob.func4() # from GrandChild itself
ob.func3() # from Child2 class
ob.func2() # from Child1 class
ob.func1() # from Parent class
```

```python
class Parent:
    def func1(self):
        print("This i fun1")

class Child1(Parent):
    def func2(self):
        print("This i fun2")

class GrandChild(Child1, Parent):
    def func4(self):
        print("This is fun4")

ob = GrandChild()
ob.func4() # from GrandChild itself
ob.func2() # from Child1 class
ob.func1() # from Parent class
```

```python
# Parent class for all employees
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_details(self):
        print("Name:", self.name)
        print("Age:", self.age)


# Common properties and methods related to employee salaries
class Salary:
    def __init__(self, salary):
        self.salary = salary

    def display_salary(self):
        print("Salary:", self.salary)


# Child classes inheriting from Person and Salary classes
class FullTimeEmployee(Person, Salary):
    def __init__(self, name, age, salary):
        # Calling the constructors of the parent classes
        Person.__init__(self, name, age)
        Salary.__init__(self, salary)


class PartTimeEmployee(Person, Salary):
    def __init__(self, name, age, salary):
        # Calling the constructors of the parent classes
        Person.__init__(self, name, age)
        Salary.__init__(self, salary)


class ContractEmployee(Person):
    def __init__(self, name, age, contract_period):
        # Calling the constructor of the parent class
        Person.__init__(self, name, age)
        self.contract_period = contract_period

    def display_contract_details(self):
        print("Contract Period:", self.contract_period)

        
# Creating objects for the classes
fte = FullTimeEmployee("John", 35, 50000)
pte = PartTimeEmployee("Jane", 25, 20000)
ce = ContractEmployee("Mike", 30, "6 months")

# Accessing methods of the classes using the objects
fte.display_details()
fte.display_salary()

pte.display_details()
pte.display_salary()

ce.display_details()
ce.display_contract_details()

```

```python
class Parent1:
    def __init__(self):
        print("In parent_1 init")

    def func(self):
        print("This i fun1")

class Parent2:
    def __init__(self):
        print("In parent_2 init")

    def func(self):
        print("This i fun2")

class Child(Parent1,Parent2):
    def __init__(self):
        super().__init__()
        print("In Child init")

    def func3(self):
        print("This is fun3")

ob = Child() # it chooses Parent1. because of (METHOD RESOLUTION ORDER (MRO))
ob.func() # when both parents have same function name.(MRO)
```

```python
class Parent1:
    def __init__(self):
        print("In parent1 init")

    def func1(self):
        print("This i fun1")

class Parent2:
    def __init__(self):
        print("In parent2 init")

    def func2(self):
        print("This i fun2")

class Child(Parent1,Parent2):
    def __init__(self):
        super().__init__()
        print("In Child init")

    def func3(self):
        super().func2()   # it also works to call functions of parent class
        print("This is fun3")

ob = Child()
ob.func3()
```

```python
# Banking System

class Account:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print("Insufficient funds.")

    def get_balance(self):
        return self.balance


class SavingsAccount(Account):
    def __init__(self, account_number, balance, interest_rate):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate

    def apply_interest(self):
        self.balance += self.balance * self.interest_rate / 100


class CheckingAccount(Account):
    def __init__(self, account_number, balance, overdraft_limit):
        super().__init__(account_number, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if amount <= self.balance + self.overdraft_limit:
            self.balance -= amount
        else:
            print("Transaction declined due to insufficient funds.")


# Create instances of different types of accounts
savings = SavingsAccount("12345", 1000, 2)
checking = CheckingAccount("67890", 500, 200)

# Perform operations on the accounts
savings.deposit(500)
savings.apply_interest()

checking.withdraw(300)
checking.withdraw(500)

# Print balances
print("Savings Account Balance:", savings.get_balance())  # Output: 1530.0
print("Checking Account Balance:", checking.get_balance())  # Output: -300.0
```
