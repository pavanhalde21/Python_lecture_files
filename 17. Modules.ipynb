# Random Module

```python
# The random module is a built-in Python module that provides a suite of functions for generating random numbers and random selections from sequences. 
# The random module can be used for a wide range of applications, such as simulations, games, cryptography, and machine learning.
```

```python
import random

# Generate a random integer between 1 and 100
random_int = random.randint(1, 6)
print(random_int)
```

```python
import random

# Generate a random floating-point number between 0 and 1

random_float = random.random()
print(random_float)


# Generate a random floating-point number between 0 and 10

random_float_2 = random.uniform(0, 10)
print(random_float_2)
```

```python
import random

# Shuffle a list of numbers

numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print(numbers)
```

```python
import random

# Select a random element from a list

fruits = ['apple', 'banana', 'cherry', 'orange']
random_fruit = random.choice(fruits)
print(random_fruit)
```

```python
#Random Passwords

import random
import string

# Define a function to generate a random password

def generate_password(length):
    characters = string.ascii_letters + string.digits + string.punctuation # string.ascii_letters, string.digits, string.punctuation
    password = ''.join(random.choice(characters) for i in range(length)) # random.choice
    return password


# Generate a random password with length 10

password = generate_password(10)
print(password)
```

```python
from string import *
from random import *
# Define a function to generate a random password

def generate_password(length):
    characters = ascii_letters + digits + punctuation # ascii_letters, digits, punctuation
    password = ''.join(choice(characters) for i in range(length)) # choice
    return password

password = generate_password(10)
print(password)
```

```python
#Random OTPs

import random
import string

def generate_otp(length):
    numbers = string.digits
    OTP = (random.choice(numbers) for i in range(length))
    for i in OTP:
        print(i,end='')
    
        
        
generate_otp(4)
```

# Packages

```python
# It is an encapsulation mechanism to group related modules into a single unit.

# Package is nothing but folder or directory which represents collection of Python modules.

# Any folder or directory contains __init__.py file,is considered as a Python package.This file can be empty.
# A package can contains sub packages also.
```

![Alt text](packages_diagram.png)

```python
# The main advantages of package statement are
# 1. We can resolve naming conflicts
# 2. We can identify our components uniquely
# 3. It improves modularity of the application
```

```python
Eg 1:

D:\Python_classes>
 |-test.py
 |-pack1
 |-module1.py
 |-__init__.py


__init__.py:
empty file

module1.py:
def f1():
print("Hello this is from module1 present in pack1")

test.py (version-1):
import pack1.module1
pack1.module1.f1()

test.py (version-2):
from pack1.module1 import f1
f1()
```
