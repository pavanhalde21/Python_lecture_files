```python
# The __init__() method / Constructor.
# Constructor is a special method in a class used to create and initialize an object of a class.

# -- The name of the constructor should be __init__(self)
# -- Constructor is invoked automatically when an object of a class is created.
# -- The main purpose of a constructor is to initialize or assign values to the data members of that class. It cannot return any value other than None.
# -- __init__() can be used to perform other operations that are necessary when the object is being created.
# -- Per object constructor will be exeucted only once.
# -- Constructor takes at least one argument(i.e. self)
# -- Constructor is optional and if we are not providing any constructor then python will provide default constructor.


########### constructor is called only once whenever an object is created
```

```python
# instance variable can be defined in:

# -- inside a constructor
# -- inside an instance(method/function)
# -- outside the class
```

```python
# first code

class Student:
    def __init__(self, name, age, grade):
        # Instance variables                  # -----> inside constructor
        self.name = name
        self.age = age
        self.grade = grade

    def get_info(self):
        self.last_name = "Smith"              # -----> inside instance (method/fuction)
        return f"{self.name} and {self.last_name} is {self.age} years old and is in grade {self.grade}."


student1 = Student("John", 12, 6) # Creating objects of Student class


print(student1.name)     
print(student1.age) 
     
print(student1.get_info()) 

student1.name = 'ajay' # we can modify name using object (this will be availble for student1 object only)

student1.school = "ZP HighSchool"             # -----> outside the class

print(student1.__dict__)
```

```python
# second code

class Student:
    def __init__(self, name, age, grade):
        # Instance variables              
        self.name = name
        self.age = age
        self.grade = grade

    def get_info(self):
        last_name = "Smith"              # -----> local variable 
        return f"{self.name} and {last_name} is {self.age} years old and is in grade {self.grade}"
```

```python
# Both codes will work fine, but the first one is more appropriate for the purpose of the get_info method, 
# which seems to be about adding a last_name attribute to the instance.

# In the first code:

# The last_name is defined as self.last_name inside the get_info method. 
# This means it becomes an instance variable and will be accessible throughout the class instance's lifetime. 
# Any other methods within the class can also access this last_name attribute, and it will be available as long as the instance exists.
# This is useful if you need to store the last name as part of the instance data and access it in other methods or parts of the class.

# In the second code:

# The last_name is defined as a local variable within the get_info method. 
# It is not associated with the instance but rather only available within the scope of the method.
# This means that the last_name will not be accessible outside the get_info method. 
# If you need to use the last name in other methods of the class or access it after calling get_info, you won't be able to do so.
# Considering the intention of the get_info method, the first code is better as it creates an instance variable last_name,
# allowing other parts of the class to access it if needed. 
# It also aligns with the idea of storing the last name as part of the student instance's data, 
# making it useful for maintaining information about the student's name.
```

```python
# You can create object of inner class inside the outer class 
# (or) You can create object of inner class outside the outer class provided you use outer class name to call it
```

```python
class Student:  # Outer class
    def __init__(self, name, rollno):
        self.name = name
        self.rollno = rollno

    def show(self):
        print(self.name, self.rollno)

    class Laptop:  # Inner class
        def __init__(self):
            self.brand = 'HP'
            self.cpu = 'i5'
            self.ram = 8
        
        def show(self):
            print(self.brand, self.cpu, self.ram)


s1 = Student("Navin",2)  # object of outer class
s2 = Student("Jenny",3)

lap1 = Student.Laptop()  # object of inner class
print(lap1.brand)
```

```python
class Student:  # Outer class
    def __init__(self, name, rollno):
        self.name = name
        self.rollno = rollno
        self.lap = self.Laptop()

    def show(self):
        print(self.name, self.rollno)
        self.lap.show()

    class Laptop:  # Inner class
        def __init__(self):
            self.brand = 'HP'
            self.cpu = 'i5'
            self.ram = 8
        
        def show(self):
            print(self.brand, self.cpu, self.ram)


s1 = Student("Navin",2)
s2 = Student("Jenny",3)

s1.show() # show() method of outer class as well as inner class both are displayed
```
